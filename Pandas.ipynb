## Concate Data frames

```python
india_weather = pd.DataFrame({"city" : ["mumbai","bangalore","Hyderabad"],
                              "Temperature":[32,45,35]})
india_weather
```

```python
US_weather = pd.DataFrame({
    "city" : ["New York","Chicago","orlando"],
    "Temperature":[32,45,35]
})
US_weather
```

```python
df = pd.concat([india_weather,US_weather],ignore_index=True)
df
```

```python
df = pd.concat([india_weather,US_weather],axis=1)
df
```

## Merge DataFrames

```python
temp_df = pd.DataFrame({
    "city":["Hyderabad","Vijayawada","Vishakapatnam","Bengaluru"],
    "temperature":[62,65,60,75]
})
temp_df
```

```python
humidity_df = pd.DataFrame({
    "city":["Hyderabad","Vijayawada","Vishakapatnam","Bengaluru"],
    "humidity":[62,65,60,75]
})
humidity_df
```

```python
df = pd.merge(temp_df,humidity_df,on="city")
df
```

```python
df = pd.merge(temp_df,humidity_df,on="city",how="outer")
df
```
